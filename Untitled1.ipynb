# Telecom X – Predicción de Cancelación de Clientes (Churn)

En este proyecto se desarrollarán modelos de Machine Learning para predecir qué clientes tienen mayor probabilidad de cancelar sus servicios.

El objetivo es ayudar a Telecom X a identificar patrones y desarrollar estrategias de retención más efectivas.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

datos = pd.read_json(url)

datos.head()
```

```python
cliente = pd.json_normalize(datos["customer"])
telefono = pd.json_normalize(datos["phone"])
internet = pd.json_normalize(datos["internet"])
cuenta = pd.json_normalize(datos["account"])
```

```python
datos_limpios = pd.concat(
    [datos["customerID"], datos["Churn"], cliente, telefono, internet, cuenta],
    axis=1
)

datos_limpios.head()
```

```python
datos_limpios.info()
```

```python
datos_limpios.describe()
```

```python
datos_limpios["Charges.Total"] = pd.to_numeric(datos_limpios["Charges.Total"], errors="coerce")
```

```python
datos_limpios["Charges.Total"] = pd.to_numeric(datos_limpios["Charges.Total"], errors="coerce")
```

```python
datos_limpios = datos_limpios.dropna()
```

```python
datos_limpios.isnull().sum()
```

```python
datos_limpios = datos_limpios.drop(columns=["customerID"])
```

#***Verificar distribución de cancelación***
 *Distribución de churn*

```python
datos_limpios["Churn"].value_counts()
```

```python
datos_limpios["Churn"].value_counts(normalize=True)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="Churn",
    data=datos_limpios,
)

plt.title("Distribución de cancelación de clientes")

plt.xlabel("Cancelación")

plt.ylabel("Cantidad de clientes")

plt.show()
```

# **Encoding**

```python
datos_modelo = pd.get_dummies(datos_limpios, drop_first=True)
```

```python
datos_modelo.head()
```

# **Análisis de correlación**

```python
#matríz de correlación
plt.figure(figsize=(12,8))

sns.heatmap(
    datos_modelo.corr(),
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Matriz de correlación entre variables")

plt.show()
```

```python
#definición  de variables
X = datos_modelo.drop("Churn_Yes", axis=1)

y = datos_modelo["Churn_Yes"]
```

# **Normalización de datos**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
#modelo 1: regresión
from sklearn.linear_model import LogisticRegression

modelo_lr = LogisticRegression(max_iter=1000)

modelo_lr.fit(X_train, y_train)

predicciones = modelo_lr.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

modelo_lr = LogisticRegression(max_iter=1000)

modelo_lr.fit(X_train, y_train)

predicciones = modelo_lr.predict(X_test)
```

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    confusion_matrix(y_test, predicciones),
    annot=True,
    fmt="d",
    cmap="Purples"
)

plt.title("Matriz de Confusión")

plt.xlabel("Predicción")

plt.ylabel("Valor real")

plt.show()
```

# **Importancia de variables**

```python
importancia = pd.Series(
    modelo_lr.coef_[0],
    index=X.columns
)

importancia.sort_values().plot(
    kind="barh",
    figsize=(10,6),
    color="#9370DB"
)

plt.title("Importancia de variables en la cancelación")

plt.xlabel("Impacto en el modelo")

plt.show()
```

## Conclusiones

El modelo de regresión logística permitió identificar variables que influyen en la cancelación de clientes.

Se observa que factores como el tipo de contrato, el tiempo de permanencia del cliente y los cargos mensuales tienen una relación importante con la probabilidad de cancelación.

Estos resultados pueden ayudar a Telecom X a desarrollar estrategias de retención enfocadas en clientes con mayor riesgo de churn.
